```python
class BinaryTree:
    def __init__(self, val):
        self.root = val
        self.left = None
        self.right = None

    def insertLeft(self, val):
        self.left = BinaryTree(val)

    def insertRight(self, val):
        self.right = BinaryTree(val)

    def getRight(self):
        return self.right

    def getLeft(self):
        return self.left

    def setRootVal(self, val):
        self.root = val

    def getRootVal(self):
        return self.root

    def __str__(self) -> str:
        def helper(self):
            if not self.getLeft() and not self.getRight():
                return [self.getRootVal(), [], []]
            elif not self.getLeft():
                return [self.getRootVal(), [], helper(self.getRight())]
            elif not self.getRight():
                return [self.getRootVal(), helper(self.getLeft()), []]
            else:
                return [
                    self.getRootVal(),
                    helper(self.getLeft()),
                    helper(self.getRight()),
                ]
        return str(helper(self))

    def height(self):
        heights = [-1]
        def iterate(self):
            if self:
                heights.append(heights[-1] + 1)
                iterate(self.getLeft())
                iterate(self.getRight())
                heights.append(heights[-1] - 1)
        iterate(self)
        return max(heights)
```

```python
def buildTree():
    tree = BinaryTree("a")
    tree.insertLeft("b")
    tree.insertRight("c")
    tree.getLeft().insertRight("d")
    tree.getRight().insertLeft("e")
    tree.getRight().insertRight("f")
    return tree

tree = buildTree()
```
