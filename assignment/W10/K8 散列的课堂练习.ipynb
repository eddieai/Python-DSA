## 【K08】散列的课堂练习
请设计算法重新实现散列表的put方法，使得散列表可以在负载因子达到一个预设值时自动调整大小。

提示：算法需要处理现有在散列表中的数据项，保证在扩充散列表之后，能正常访问现有的数据项。

可以采用伪代码说明算法。

提示：
首先建立一个长度为n的散列表。我们不妨认为将散列函数为对n取余数，将已有的数据项逐个填入该散列表。采用二次探测解决冲突。
当填入散列表的负载因子达到预设值的时候，生成一个新的散列表。新的散列表的长度为原来散列表的l倍，我们不妨假设这个l的值为2，则这个散列表的长度为2n。
我们将原来的散列表中的数据值填入新的散列表，一一对应填入新的散列表的前n个位置。
对于新的数据值，我们采用的散列函数为对n取余加n，将新的数据填入新的散列表。
