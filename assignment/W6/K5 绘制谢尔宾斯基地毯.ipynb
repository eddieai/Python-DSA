```python
import turtle


def sierpinskiSquare(top, left, bottom, right, degree):

    drawSquare(top, left, bottom, right)

    if degree > 0:
        hole_size = (bottom - top) // 3
        for row in range(3):
            for col in range(3):
                if not (row == 1 and col == 1):
                    sierpinskiSquare(
                        top + row * hole_size,
                        left + col * hole_size,
                        top + (row + 1) * hole_size,
                        left + (col + 1) * hole_size,
                        degree=degree - 1
                    )


def drawSquare(top, left, bottom, right):  # 
    t.fillcolor("white")
    t.pencolor("blue")
    t.pensize(2)
    t.penup()
    t.goto((left, top))
    t.pendown()
    t.begin_fill()
    t.goto((left, bottom))
    t.goto((right, bottom))
    t.goto((right, top))
    t.goto((left, top))
    t.end_fill()


t = turtle.Turtle()
turtle.tracer(0, 0)
sierpinskiSquare(300, 300, -300, -300, degree=4)
turtle.update()
turtle.done()
```
