## 后缀表达式求值的栈变化
- 要求每次push/pop后都需要显示栈中的元素

```python
# 后缀表达式求值
class Stack(list):
    def push(self, item):
        self.append(item)
        print(self)
    def pop(self):
        tmp = super().pop()
        print(self)
        return tmp
    def peek(self):
        return self[-1]
    def isEmpty(self):
        return self == []
    def size(self):
        return len(self)
    def __str__(self):
        res = ""
        for item in self:
            res += f"| {item} "
        return res[1:]


def postfixEval(postfixExpr):
    postfixExpr = postfixExpr.split(" ")
    stack = Stack()
    for token in postfixExpr:
        if "0" <= token <= "9":
            stack.push(token)
        else:
            num2 = stack.pop()
            num1 = stack.pop()
            newnum = eval(f"{num1}{token}{num2}")
            stack.push(newnum)
    return stack.pop()
```

```python
print(postfixEval('2 3 * 4 +'))
```

```python
print(postfixEval('1 2 + 3 + 4 + 5 +'))
```

```python
print(postfixEval('1 2 3 4 5 * + * +'))
```
