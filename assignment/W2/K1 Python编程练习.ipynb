## 下面编写程序

```python
# 读取BCD格式数据文件
# 给出输入提示，输入开始位置和数字数量
# 打印输出读取的数字序列

from time import time

with open("../data/pi50.4.bin", "rb") as f:
    pi50 = f.read()

def pisearch(s, l):
    t0 = time()
    s -= 1
    e = s + l
    s_half = s // 2
    e_half = e // 2 + e % 2

    res = "".join([str(hex(a))[-2:] for a in pi50[s_half:e_half]]).replace("x", "0")

    s_ = 1 if s%2 else None
    e_ = -1 if e%2 else None

    print(res[s_:e_])
    print(time()-t0)
```

```python
pisearch(20210316, 25)
pisearch(20200101, 10)
```
