## 【H1】在Pi中查找生日
请在提供的“`pi50.4.bin`”文件所包含的5000万位圆周率中查找日期，编写程序实现如下功能：
1. 输入一个8位数字的日期，打印这个日期所在的位置，如果找不到，则输出“`NOT FOUND`”；
2. 输入一个4位数字的年份，如“`2024`”，查找从`20240101`到`20241231`的全年中，每个日期是否存在于5000万位圆周率中；要求输出3个数值:(存在的日期个数；不存在的日期个数；查找所花的总时间（秒）)


## 回答

### 题目1

```python
with open("..\data\pi50.4.bin", "rb") as f:
    pi50 = f.read()
pi50 = "".join([str(hex(a))[-2:] for a in pi50]).replace("x", "0").encode()
```

```python
def pisearch_date(date):
    if ind:=pi50.find(date.encode()) != -1:
        print(ind)
    else:
        print("NOT FOUND")
```

```python
pisearch_date("19890209")
```

### 题目2

```python
from time import time
from datetime import datetime, timedelta


def pisearch_2022_dates():
    t0 = time()

    start_date = datetime(2022, 1, 1)
    end_date = datetime(2023, 1, 1)
    all_dates = set()
    current_date = start_date
    while current_date < end_date:
        formatted_date = current_date.strftime("%Y%m%d")
        all_dates.add(formatted_date.encode())
        current_date += timedelta(days=1)

    exist = 0
    i = 0
    while True:
        i = pi50.find("2022".encode(), i+1)
        if i == -1:
            break
        if pi50[i:i+8] in all_dates:
            exist += 1
            all_dates.remove(pi50[i:i+8])

    print(exist, 365-exist, time() - t0)
```

```python
pisearch_2022_dates()
```
