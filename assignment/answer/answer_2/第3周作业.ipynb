# 1. 有效的括号（10分）

**题目内容：**

给定一个只包括 `'('`，`')'`，`'{'`，`'}'`，`'['`，`']'` 的字符串，判断字符串是否有效。<br>
有效字符串需满足：<br>
左括号必须用相同类型的右括号闭合。<br>
左括号必须以正确的顺序闭合。<br>
注意空字符串可被认为是有效字符串。



**输入格式:**

一行字符串



**输出格式：**

True或False，表示该输入是否为合法括号串



**输入样例：**

`([])`



**输出样例：**

`True`



参考代码模板：

```python
def isValid(s):
    ### code here
 
print(isValid(input()))
```

时间限制：500ms  内存限制：32000kb

````python
def isValid(s):
    left = '([{'
    right = ')]}'
    left_stack = []
    for c in s:
        if c in left:
            left_stack.append(c)
        else:
            if left_stack and left_stack[-1] == left[right.index(c)]:
                left_stack.pop()
            else:
                return False
    if left_stack:
        return False
    return True
        

print(isValid(input()))
````

# 2. 每日温度（10分）

**题目内容：**

根据每日气温列表，请重新生成一个列表，对应位置的输入是你需要再等待多久温度才会升高超过该日的天数。如果之后都不会升高，请在该位置用 0 来代替



**输入格式:**

一行以Python表达式格式给出的列表，包含数个整数



**输出格式：**

整数组成的列表，直接使用print输出



**输入样例：**

`[73, 74, 75, 71, 69, 72, 76, 73]`



**输出样例：**

`[1, 1, 4, 2, 1, 1, 0, 0]`



**参考代码模板：**

```python
def dailyTemp(T):
    ### code here
 
t = eval(input())
print(dailyTemp(t))
```

时间限制：500ms  内存限制：32000kb

````python
def dailyTemp(T):
    T_stack = T[::-1]
    temp_stack = []
    result = []
    while len(T_stack) > 1:
        t_now = T_stack.pop()
        print(t_now)
        while T_stack:
            t_future = T_stack.pop()
            temp_stack.append(t_future)
            if t_future > t_now:
                result.append(len(temp_stack))
                break
        else:
            result.append(0)
        while temp_stack:
            T_stack.append(temp_stack.pop())
    result.append(0)
    return result
                

t = eval(input())
print(dailyTemp(t))
````

# 3. 后缀表达式求值（10分）

**题目内容：**

根据后缀表达式表示法，求表达式的值。<br>
有效的运算符包括 +, -, *, / ；其中除法仅保留整数结果。<br>
若出现除数为0或表达式非法的情况，输出"NaN"<br>
注：
1. 每个数字token可包含多位，也可为负数
2. 除法统一使用 int(a / b) 而非 a // b 进行计算



**输入格式:**

一行字符串，每个token间以空格分隔



**输出格式：**

一行，包含一个整数或"NaN"



**输入样例：**

`"4 13 5 / +"`



**输出样例：**

`6`



**示例代码模板：**

```python
def calc(tokens):
    ### code here
 
print(calc(input().split()))
```

时间限制：500ms  内存限制：32000kb

````python
def calc(tokens):
    stack = []
    ops = '+-*/'
    for token in tokens:
        if token not in ops:
            stack.append(eval(token))
        elif len(stack) > 1:
            num_right = stack.pop()
            num_left = stack.pop()
            op = ops.index(token)
            if op == 0:
                stack.append(num_left + num_right)
            if op == 1:
                stack.append(num_left - num_right)
            if op == 2:
                stack.append(num_left * num_right)
            if op == 3:
                if num_right:
                    stack.append(int(num_left / num_right))
                else:
                    return 'NaN'
        else:
            return 'NaN'
    if len(stack) != 1:
        return 'NaN'
    else:
        return stack.pop()

print(calc(input().split()))
````
